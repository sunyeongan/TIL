- Matrix operations

    - sum(axis)
    - 행렬 더하기가 성립하려면 차원이 같아야 한다.

```python
C = np.random.randn(3,3)
C
```

```python
D = np.random.randn(3,3)
D
```

```python
C+D
```

```python
C-D
```

```python
c1 = C[0,:]
d1 = D[0,:]
c1+d1
```

```python
a3 = np.array([1,2,3,4])
b3 = 2*np.ones(4)
b3
```

```python
a3
```

```python
c3 = a3*b3
c3
```

```python
np.sum(c3)#c에 있는 원소들을 모두 더해라
```

```python
np.matmul(a3,b3.reshape(-1,1))
```

```python
a4 = np.arange(3,11,1).reshape(2,4)
a4
```

```python
np.sum(a4, axis=0)#위아래로 더함(열에 있는 원소를 모두 더함)
```

```python
np.sum(a4,axis=1)#행에 있는 원소를 모두 더함
```

```python
c4 = np.sum(a4,axis=1).reshape(-1,1)
c4
```
